```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
pd.set_option('display.max_columns', None)
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
train.head()
```

```python
train.info()
```

DataType

```python
dtype = train.dtypes
unique = train.nunique()
pd.DataFrame({'dtype': dtype, 'isnull': unique}).T
```

```python
def datatype(x):
    CatColumns = x.select_dtypes('object').columns
    x[CatColumns] = x[CatColumns].astype('category')
```

```python
datatype(train)
```

```python
datatype(test)
```

Missing Values

```python
null = train.isnull().sum()
ratio = 100*null / train.shape[0]
pd.DataFrame({'Null_Sum': null, 'Ratio': ratio}).T
```

```python
train = train.ffill()
```

```python
test = test.ffill()
```

```python
train.drop(['Id', 'Alley', 'MasVnrType', 'FireplaceQu', 'PoolQC', 'MiscFeature', 'Fence'], axis=1, inplace=True) 
test.drop(['Alley', 'MasVnrType', 'FireplaceQu', 'PoolQC', 'MiscFeature', 'Fence'], axis=1, inplace=True)
```

```python
null = train.isnull().sum()
ratio = 100*null / train.shape[0]
pd.DataFrame({'Null_Sum': null, 'Ratio': ratio}).T
```

Encoding

```python
def encoding(x):
    CatCols = x.select_dtypes('category').columns
    for col in CatCols:
        x[col] = x[col].astype('category').cat.codes
```

```python
encoding(train)
```

```python
encoding(test)
```

```python
drop_col = ['MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'LowQualFinSF', 'KitchenQual', 'Functional', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'SaleType', 'SaleCondition']
train.drop(drop_col, axis=1, inplace=True)
test.drop(drop_col, axis=1, inplace=True)
```

MultiRegression Model

```python
X = train.drop(['SalePrice'], axis=1)
y = train['SalePrice']
```

```python
print(X.shape)
print(y.shape)
```

```python
y_log = np.log1p(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_log, test_size=0.3, random_state=0)
```

```python
#model = LinearRegression().fit(X_train, y_train)
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42).fit(X_train, y_train)
```

```python
y_pred_log = model.predict(X_test)
```

```python
RMSE_log = np.sqrt(mean_squared_error(y_test, y_pred_log))
print(f"RMSE (log-transformed): {RMSE_log}")
```

```python
y_pred_original = np.expm1(y_pred_log)
```

```python
print(f"R^2 score: {model.score(X_test, y_test)}")
```
